```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('rentalOffices.csv', index_col=None)
df1.drop("Unnamed: 0", axis=1, inplace=True)
df2 = pd.read_csv('rentalOffices1.csv', index_col=None)
df2.drop("Unnamed: 0", axis=1, inplace=True)
df3 = pd.read_csv('rentalOffices2.csv', index_col=None)
df3.drop("Unnamed: 0", axis=1, inplace=True)
df4 = pd.read_csv('rentalOfficesCopy.csv', index_col=None)
df4.drop("Unnamed: 0", axis=1, inplace=True)
df5 = pd.read_csv('rentalOfficesLast.csv', index_col=None)
df5.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
dfs = [df1, df2, df3, df4, df5]
merged_df = pd.concat(dfs, ignore_index=True)

def merge_duplicates(group):
    return group.fillna(method="ffill").fillna(method="bfill").iloc[0]

final_df = (
    merged_df.groupby("Link", dropna=False)
    .apply(merge_duplicates)
    .reset_index(drop=True)
)

final_df.info()
```

```python
import pandas as pd
import re
from dateutil.relativedelta import relativedelta
from word2number import w2n

today = pd.to_datetime("2025-01-18")


def convert_to_actual_date(relative_str, today):
    relative_str = relative_str.lower()

    patterns = {
        r"(\d+)\s*second[s]?\s*ago": lambda m: today - relativedelta(seconds=int(m.group(1))),
        r"(\d+)\s*minute[s]?\s*ago": lambda m: today - relativedelta(minutes=int(m.group(1))),
        r"(\d+)\s*hour[s]?\s*ago": lambda m: today - relativedelta(hours=int(m.group(1))),
        r"(\d+)\s*day[s]?\s*ago": lambda m: today - relativedelta(days=int(m.group(1))),
        r"(\d+)\s*week[s]?\s*ago": lambda m: today - relativedelta(weeks=int(m.group(1))),
        r"(\d+)\s*month[s]?\s*ago": lambda m: today - relativedelta(months=int(m.group(1))),
        r"(\d+)\s*year[s]?\s*ago": lambda m: today - relativedelta(years=int(m.group(1))),
    }

    for pattern, func in patterns.items():
        match = re.search(pattern, relative_str)
        if match:
            return func(match)

    return None

final_df['PostingDate'] = final_df['PostingDate'].apply(lambda x: convert_to_actual_date(x, today))

final_df.head()
```

```python
final_df.to_csv('NewCairoRentalOffices.csv', index=False)
```
